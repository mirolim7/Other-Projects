## Q1 (30%)
Use Credit.csv dataset. Do the following tasks:
1. Split your dataset to train and test sets (20% for testing) with random state equal to 2021. Imagine that you can use test set only for evaluation at the end.
2. Using any model of your choice, predict Balance. To evaluate performance of the trained model, obtain MSE of training set, 5-fold CV, 10-fold CV, and test set. Present all results in the form of tables (dataframes).
3. Instead of using CV, using any random split obtain validation set of the same size as the test set. Train the model on training part, and evaluate performance of the trained model on training, validation and test sets. Present all results in the form of tables (dataframes).
4. Now, repeat the step above 5 times with 5 different random splits. You can show results in table. 
5. Comment the results obtained using CV and validation sets.


*Note: You are allowed to use sklearn, but not statsmodels.*

*Hint: To get validation set, you can use train_test_split function two times (if you use it properly).*

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import sklearn.linear_model as skl_lm
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
```

```python
df_1 = pd.read_csv('data/Credit.csv')
df_1.head()
```

### Step 1

```python
X = df_1.drop('Balance', axis=1)
y = df_1.Balance
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2021)
print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)
```

### Step 2
For simplicity, use only numerical variables. Hence, we will drop Gender, Student, Married, Ethnicity columns. Alternatively, we can convert them to dummies. 

```python
X_train.drop(['Gender', 'Student', 'Married', 'Ethnicity'], axis=1, inplace=True)
X_test.drop(['Gender', 'Student', 'Married', 'Ethnicity'], axis=1, inplace=True)

print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)
```

```python
# Dictionary to store results
mse_1 = dict()

# Define model
model_lr = skl_lm.LinearRegression()

# Cross Validation with 5 folds
mse_1['5-fold CV'] = cross_val_score(model_lr, X_train, y_train, cv = 5, 
                                             scoring='neg_mean_squared_error').mean() * (-1)

# Cross Validation with 10 folds
mse_1['10-fold CV'] = cross_val_score(model_lr, X_train, y_train, cv = 10, 
                                             scoring='neg_mean_squared_error').mean() * (-1)

# Fit the model
model_lr.fit(X_train, y_train)

# Get Train MSE
mse_1['Train'] = mean_squared_error(y_train, model_lr.predict(X_train)) 

# Get Test MSE
mse_1['Test'] = mean_squared_error(y_test, model_lr.predict(X_test))

mse_1
```

```python
# Save results in dataframe 
results_1_1 = pd.DataFrame({'MSE': mse_1.values()}, index=mse_1.keys())
results_1_1
```

### Step 3

```python
X_train_1, X_val_1, y_train_1, y_val_1 = train_test_split(X_train, y_train, test_size=80, random_state=1)
print(X_train_1.shape)
print(y_train_1.shape)

print(X_val_1.shape)
print(y_val_1.shape)
```

```python
# Dictionary to store results
train_mse = dict()
val_mse = dict()
test_mse = dict()

# Define and fit model
model_lr_2 = skl_lm.LinearRegression().fit(X_train_1, y_train_1)

# Get Train MSE
train_mse['State: 1'] = mean_squared_error(y_train_1, model_lr_2.predict(X_train_1)) 

# Get Validation MSE
val_mse['State: 1'] = mean_squared_error(y_val_1, model_lr_2.predict(X_val_1)) 

# Get Test MSE
test_mse['State: 1'] = mean_squared_error(y_test, model_lr_2.predict(X_test))

# Save results in dataframe 
results_1_2 = pd.DataFrame({'Train': train_mse.values(), 
                          'Validation': val_mse.values(), 
                          'Test': test_mse.values()}, index=train_mse.keys())
results_1_2
```

### Step 4

```python
# Dictionary to store results
train_mse = dict()
val_mse = dict()
test_mse = dict()

for state in range(1,7):
    X_train_, X_val_, y_train_, y_val_ = train_test_split(X_train, y_train, test_size=0.25, random_state=state)
    # Define and fit model
    model = skl_lm.LinearRegression().fit(X_train_, y_train_)

    # Get Train MSE
    train_mse['State: ' + str(state)] = mean_squared_error(y_train_, model.predict(X_train_)) 

    # Get Validation MSE
    val_mse['State: ' + str(state)] = mean_squared_error(y_val_, model.predict(X_val_)) 

    # Get Test MSE
    test_mse['State: ' + str(state)] = mean_squared_error(y_test, model.predict(X_test))

# Save results in dataframe 
results_1_3 = pd.DataFrame({'Train': train_mse.values(), 
                            'Validation': val_mse.values(), 
                            'Test': test_mse.values()}, index=train_mse.keys())
results_1_3
```

### Step 5
There is a high fluctuation in MSE when we use Validation set, while it is more stable with Cross Validation.

## Q2 (20%)
Use boston.csv dataset. Do the following tasks: 
1. Split your dataset to train and test sets (20% for testing) with random state equal to 2021. Imagine that you can use test set only for evaluation at the end.

2. Your task is to predict target variable - **medv**.

3. Use Ridge regression with alpha (penalty term) varying from 0 to some large value of alpha, at which all coefficients take value of 0. Using Cross-Validation select the most optimal value of alpha. 

4. For the selected value of alpha train your model on the training set (80%). And then evaluate on test set (20%). 

5. Show the coefficients of the final model.

*Note: You should use regular Ridge model from sklearn (not RidgeCV) and apply CV on it.*

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
```

```python
df_2 = pd.read_csv('data/boston.csv')
df_2.head()
```

```python
print(df_2.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_2.drop(['medv'], axis=1), df_2['medv'],
                                                    test_size=0.2, random_state=2021)
print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)
```

```python
# Define scaler for X data
scaler = StandardScaler().fit(X_train)

# Define model
model_ridge = Ridge()

# Dictionary to store results
ridge_cv = dict()
```

```python
alphas = 10**np.linspace(10,-2,500)*0.5

for a in alphas:
    model_ridge.set_params(alpha=a)
    ridge_cv[a] = cross_val_score(model_ridge, scaler.transform(X_train), y_train, 
                                  cv=5, scoring='neg_mean_squared_error').mean() * (-1)
```

```python
ridge_cv
```

```python
results_2_1 = pd.DataFrame({'Alpha': ridge_cv.keys(), 'MSE': ridge_cv.values()})
results_2_1.sort_values(by='MSE', ascending=True, inplace=True)
results_2_1.reset_index(drop=True, inplace=True)
results_2_1
```

```python
optimal_alpha = results_2_1.Alpha[0]
optimal_alpha
```

```python
# Set up model with the optimal alpha
model_ridge.set_params(alpha=optimal_alpha)

# Fit model on full training data
model_ridge.fit(scaler.transform(X_train), y_train)

# Dictionary to store results
mse_2 = dict()

# Get Train MSE
mse_2['Train'] = mean_squared_error(y_train, model_ridge.predict(scaler.transform(X_train))) 

# Get Test MSE
mse_2['Test'] = mean_squared_error(y_test, model_ridge.predict(scaler.transform(X_test)))

# Save results in dataframe 
results_2_2 = pd.DataFrame({'MSE': mse_2.values()}, index=mse_2.keys())
results_2_2
```

```python
print('Intercept:', round(model_ridge.intercept_,3))
```

```python
coefs = pd.DataFrame({'Parameters': X_train.columns, 'Coefficients':model_ridge.coef_})
coefs
```

Again use boston.csv dataset. Do the following tasks: 
1. Split your dataset to train and test sets (20% for testing) with random state equal to 2021. Imagine that you can use test set only for evaluation at the end.

2. Again, your task is to predict target variable - **medv**.

3. Use Lasso regression with alpha (penalty term) varying from 0 to some large value of alpha, at which all coefficients take value of 0. Using Cross-Validation select the most optimal value of alpha. 

4. For the selected value of alpha train your model on the training set (80%). And then evaluate on test set (20%). 

5. Show the coefficients of the final model.

6. Compare your results (i.e. test MSE) and coefficients with the Ridge Model.

*Note: In this case you should use LassoCV (not Lasso and apply CV separately).*

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LassoCV
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
df_2 = pd.read_csv('data/boston.csv')
df_2.head()
```

```python
print(df_2.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_2.drop(['medv'], axis=1), df_2['medv'],
                                                    test_size=0.2, random_state=2021)
print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)
```

```python
# Define scaler for X data
scaler = StandardScaler().fit(X_train)

alphas = 10**np.linspace(10,-2,500)*0.5
```

```python
print(alphas[:5])
print(alphas[-5:])
```

```python
# Define model
lassocv = LassoCV(alphas=alphas, cv=5)

# Fit model
lassocv.fit(scaler.transform(X_train), y_train)
```

```python
results_3_1 = pd.DataFrame({'Alpha': lassocv.alphas_, 'MSE': lassocv.mse_path_.mean(axis=1)})
results_3_1.sort_values(by='MSE', ascending=True, inplace=True)
results_3_1.reset_index(drop=True, inplace=True)
results_3_1
```

```python
# Using method alpha_ returns alpha with the lowest MSE, i.e. the most optimal alpha 
optimal_alpha = lassocv.alpha_
optimal_alpha
```

```python
# Dictionary to store results
mse_3 = dict()

# Get Train MSE
mse_3['Train'] = mean_squared_error(y_train, lassocv.predict(scaler.transform(X_train))) 

# Get Test MSE
mse_3['Test'] = mean_squared_error(y_test, lassocv.predict(scaler.transform(X_test)))


# Save results in dataframe 
results_3_2 = pd.DataFrame({'MSE': mse_3.values()}, index=mse_3.keys())
results_3_2
```

```python
print('Intercept:', round(lassocv.intercept_,3))
```

```python
coefs = pd.DataFrame({'Parameters': X_train.columns, 'Coefficients':lassocv.coef_})
coefs
```

## Consider LASSO separately with optimal alpha

```python
model_lasso = Lasso().set_params(alpha=optimal_alpha)
model_lasso.fit(scaler.transform(X_train), y_train)

# Dictionary to store results
mse_4 = dict()

# Get Train MSE
mse_4['Train'] = mean_squared_error(y_train, model_lasso.predict(scaler.transform(X_train))) 

# Get Test MSE
mse_4['Test'] = mean_squared_error(y_test, model_lasso.predict(scaler.transform(X_test)))


# Save results in dataframe 
results_3_3 = pd.DataFrame({'MSE': mse_4.values()}, index=mse_4.keys())
results_3_3
```

Use Wage.csv dataset. Do the following tasks:

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, BaggingRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
```

```python
df_4 = pd.read_csv('data/Wage.csv')
df_4.head()
```

```python
df_4.shape
```

```python
df_4.nunique()
```

```python
df_4.isnull().sum()
```

```python
df_4.sex.unique()
```

```python
df_4.region.unique()
```

```python
# We can drop these columns for sure
df_4.drop(['Unnamed: 0', 'logwage', 'sex', 'region'], axis=1, inplace=True)
```

```python
# Preprocessing: convert to dummies
df_4.year = pd.factorize(df_4.year)[0]
df_4.maritl = pd.factorize(df_4.maritl)[0]
df_4.race = pd.factorize(df_4.race)[0]
df_4.education = pd.factorize(df_4.education)[0]
df_4.jobclass = pd.factorize(df_4.jobclass)[0]
df_4.health = pd.factorize(df_4.health)[0]
df_4.health_ins = pd.factorize(df_4.health_ins)[0]
df_4
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_4.drop(['wage'], axis=1), df_4['wage'],
                                                    test_size=0.2, random_state=2021)
print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)
```

```python
model_tree = DecisionTreeRegressor()
model_tree.fit(X_train, y_train)

print(f'MSE train: {mean_squared_error(y_train, model_tree.predict(X_train))}')
print(f'MSE test: {mean_squared_error(y_test, model_tree.predict(X_test))}')
```

```python
model_bag = BaggingRegressor()
model_bag.fit(X_train, y_train)

print(f'Bagging MSE train: {mean_squared_error(y_train, model_bag.predict(X_train))}')
print(f'Bagging MSE test: {mean_squared_error(y_test, model_bag.predict(X_test))}')
```

```python
model_rf = RandomForestRegressor()
model_rf.fit(X_train, y_train)

print(f'Random Forest MSE train: {mean_squared_error(y_train, model_rf.predict(X_train))}')
print(f'Random Forest MSE test: {mean_squared_error(y_test, model_rf.predict(X_test))}')
```

### Comment
Even under default settings of the models, we can see drammatic improvement in performance of the model on Test set. Also, note that Train MSE increased a bit.

```python
# Let's consider n_estimators parameter in the model, which is the number of trees in the forest.
```

```python
N = np.arange(25,251,25)
N
```

```python
rf_cv = dict()

for size in N:
    model = RandomForestRegressor(n_estimators=size).fit(X_train, y_train)
    rf_cv[size] = cross_val_score(model, X_train, y_train, 
                                  cv=5, scoring='neg_mean_squared_error').mean() * (-1)
```

```python
results_4_1 = pd.DataFrame({'Size': rf_cv.keys(), 'MSE': rf_cv.values()})
results_4_1.sort_values(by='MSE', ascending=True, inplace=True)
results_4_1.reset_index(drop=True, inplace=True)
results_4_1
```

```python
n_optimal = results_4_1.Size[0]
n_optimal
```

```python
model_rf = RandomForestRegressor(n_estimators=n_optimal)
model_rf.fit(X_train, y_train)

print(f'Random Forest MSE train: {mean_squared_error(y_train, model_rf.predict(X_train))}')
print(f'Random Forest MSE test: {mean_squared_error(y_test, model_rf.predict(X_test))}')
```
